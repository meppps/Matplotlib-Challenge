```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouseData = pd.read_csv(mouse_drug_data_to_load)
clinData = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mdf = pd.merge(mouseData,clinData,how="left")


# Display the data table for preview
mdf
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survivors = mdf.groupby(['Drug','Timepoint'])
# Convert to DataFrame
survivors = survivors.count().reset_index().rename(columns={"Mouse ID":"Mouse Count"}).drop(columns=['Metastatic Sites','Tumor Volume (mm3)'])


# Preview DataFrame
survivors.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survTable = survivors.pivot_table(index='Timepoint', columns='Drug', values='Mouse Count').head(10)
# Preview the Data Frame
survTable.head(5)
```

```python
# Generate the Plot (Accounting for percentages)
y_axis=[survTable[['Capomulin']]]
x_axis=[survTable.index.values]

plt.plot(survTable[['Capomulin','Infubinol','Ketapril','Placebo']]/25*100)
plt.xlabel("Survival During Treatment")
plt.ylabel("Survival Rate %")
plt.legend(loc="best", labels=['Capomulin','Infubinol','Ketapril','Placebo'])
plt.grid()

# Save the Figure
plt.savefig("Survival.png")


# Show the Figure
plt.show()
```

## Summary Bar Graph
